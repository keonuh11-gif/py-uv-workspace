count는 특정 문자열이 몇 번 나오는지 센다.

```python
a = "hello world"
a.count('l')  # l이 3개가 들어가기 때문에 결과값은 3

# 언제 사용할까?
# 문자 개수 세기
# 특정 단어가 몇 번 나왔는지 확인

# 예시

msg = "banana"

print(msg.count('a'))

# 결과

3
```

find는 찾는 문자열이 처음 나오는 위치(인덱스)를 알려준다.

```python
s.find("world")  # world가 hello world에서 6번째부터 시작하기 때문이야
```

```python
# 찾는 문자열이 없으면?

s.find("python") # -1은 없다는 뜻
```

```python
# 언제 사용할까?
# 예를 들어 사용자가 입력한 문장에
# python
# 이라는 단어가 있는지 확인할 때.

text = input()

if text.find("python") != -1:
    print("있다.")
```
