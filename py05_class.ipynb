````python
class Person:
    def __init__(self, name, age, job):
      self.name = name  # self.멤버변수 = 매개변수
      self.age = age
      self.job = job


    def eat(self):
        print('eat')


    def walk(self):
        print('walk')
````

````python
class FourCal:


  def __init__(self, a, b):
    print('init')
    self.a = a
    self.b = b


  def add(self):
    # 클래스의 멤버변수를 메소드를 호출해서 사용할 수 있다.
    return self.a + self.b


  def sub(self):
    return self.a - self.b


  def mul(self):
    return self.a * self.b
  def div(self):
    return self.a // self.b
````

````python
fc = FourCal(4, 2)  # 생성자는 객체의 초기값을 할당한다.
````

````python
print(fc.add())
print(fc.sub())
print(fc.mul())
print(fc.div())
````

````python
def x(fun):
    def y(x):
        return fun(x+1)
    return y
````

````python
# @를 붙이면 decorator이다.
@x
def s(a):
    return a
````

````python
# s()할수에 @x 데코레이터가 선언되어 있지 않을 때 호출하는 방법
x(s)(5)

# s()함수에 @x 데코레이터가 선언되어 있을 때 호출하는 방법
s(5)
````

### classmethod
- 첫 번째 인자로 클래스(cls)를 받는다.
- 클래스를 이용해 무언가를 해야할때 사용한다.
- 클래스 변수에 접근하거나 다른 클래스 메소드를 호출할 수 있다.
- 일반적으로 팩토리 메서드 등에 자주 사용된다.


````python
class B:
  x=3          # 클래스 변수
  @classmethod
  def aa(cls):  # classmethod는 관례상 cls를 사용한다.
    print('python')


  def bb(self):
    print('instance')
````

````python
B.x  # 클래스.클래스변수
````

````python
B.aa()
````

````python
B.bb()
````

````python
ab = B()
B.bb(ab)
````

````python
class B:
    x = 1

    @classmethod
    def aa(cls, x):
        cls.x = x
````

````python
B.x
````

````python
b = B()
````

````python
b.x  # b인스턴스에 x가 없으므로 B클래스의 x를 가져온다.
````

````python
b.x = 5 # b인스턴스에 x변수를 생성한다.
````

### staticmethod
- 첫 번째 인자로 self, cls를 받지 않는다.
- 클래스나 인스턴스와 관계없는 일반적인 유틸리티 함수를 정의할때 사용한다.
- 독립적인 기능이지만 클래스 안에 넣는 것이 구조적으로 더 나을때 사용한다.


| 항목         | classmethod             | staticmethod |
| ---------- | ----------------------- | ------------ |
| 첫 번째 인자    | `cls` (클래스)             | 없음           |
| 클래스 변수 접근  | 가능                      | 불가능          |
| 인스턴스 변수 접근 | 불가능                     | 불가능          |
| 주 용도       | 클래스 상태를 사용하는 메서드        | 유틸리티 함수 정의   |
| 호출 방식      | 클래스.메서드() 또는 인스턴스.메서드() | 동일           |


````python
class B:
  x = 1


  def xx(self, x): # 인스턴스 메소드
    self.x = x


  @classmethod
  def yy(cls, x):  # 클래스 메소드
    cls.x = x


  @staticmethod
  def zz(a, b):  # 스테이틱 메소드
    print(a+b)


  @staticmethod  # 스테이틱 메소드
  def kk():
    print(x)
````

````python
B.yy(10)
````

````python
B.x
````

````python
B.zz(3,4)
````

````python
B.kk()
````

예제

피자의 반지름을 10, 도넛의 반지름은 2로 한다.

[실행결과]
피자의 면적은 314.0
도넛의 면적은 12.56

````python
class Circle:
    def setArea(self, name, radius):
        self.name = name
        self.radius = radius

    def getArea(self):
        return 3.14 * self.radius * self.radius
````

````python
ca = Circle()
ca.setArea('피자', 10)


cb = Circle()
cb.setArea('도너츠', 2)


print('{}의 면적은  {}입니다.'.format(ca.name, ca.getArea()))
print('{}의 면적은  {}입니다.'.format(cb.name, cb.getArea()))
````

```
[출력결과]
파이썬의정석의 저자는 김파이썬이고 150페이지로 구성됨
책이름 :  파이썬 입문
책저자 :  박북스
페이지 :  200 페이지
파이썬 입문의 저자는 박북스이고 200페이지로 구성됨
```

````python
class Book:

    def __init__(self, title=None, author=None, page=None):
        self.__title = title
        self.__author = author
        self.__page = page
    
    def getTitle(self):
        return self.__title
    

    def setTitle(self, title):
        self.__title = title


    def getAuthor(self):
        return self.__author


    def setAuthor(self, author):
        self.__author = author


    def getPage(self):
        return self.__page


    def setPage(self, page):
        self.__page = page


    def toString(self):
        return f"{self.__title}의 저자는 {self.__author}이고 {self.__page}페이지로 구성됨"
````

````python
class BookRun:
    def process(self):
        bk1 = Book()
        bk1.setTitle('파이썬의 정석')
        bk1.setAuthor('김파이썬')
        bk1.setPage(150)
        print(bk1.toString())


if __name__ == "__main__":
    bn = BookRun()
    bn.process()
````
