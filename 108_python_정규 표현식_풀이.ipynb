문제1] 이름과 년도을 출력하도록 구현

['(이동민, 2019)', '(최재영, 2019)', '(Lion, 2018)']

문제12] words리스트에 저장된 여러 단어중 'a'로 시작하는 단어만 출력  

apple  
arise  
above  

```python
words = ['apple', 'cat', 'brave', 'drama', 'arise', 'blow', 'coat', 'above']
```

문제14] 숫자에서  천단위를 제거


['123강남', 151767.0, 11093.0, 27394.0]

```python
data = ['123강남', '151,767', '11,093', '27,394']
```

```python
k = []

for j in data:
    if re.search('[a-z가-힣]', j):
        k.append(j)
    else:
        k.append(str(float(re.sub(',', '', j))))

k
```

문제14] 이메일 형식 체크  
  <조건>   
  1 min@daum.net  
  2 min@seoul.go.kr  



```python
email = input('이메일:')
r = re.compile('^[a-zA-Z][\w]*@[a-z]+.[a-zA-Z0-9]+[.[a-zA-Z]{2,3}]?$')
m = re.search(r, email)
print(m)
```

문제15] 아이디 정규 표현식
   <조건>       
   1. 전체 길이 5~8자   
   2. 영문자로 시작     
   3. 숫자가 2개이상 포함되여 있어야함  
   4. 2자이상의 반복문자가 연속 2번 있으면 안됨  



```python
id = input('아이디:')
r = re.compile(r'^(?=.*\d+.*\d+)(?!.*(\w{2})\1)[a-zA-Z][a-zA-Z0-9]{4,7}$')
m = re.search(r, id)
print(m)
if m:
  print('ok ')
else:
  print('fail')
```
